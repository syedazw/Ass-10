## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
# 1st method

gender_salary = adult_data.groupby('gender')['income'].mean().reset_index()
gender_salary
```

```python
# 2nd method

total_males = adult_data[adult_data['gender'] == 'Male'].shape[0]
total_females = adult_data[adult_data['gender'] == 'Female'].shape[0]

# Males and females with income >50k
males_above_50k = adult_data[(adult_data['gender'] == 'Male') & (adult_data['income'] == 1)].shape[0]
females_above_50k = adult_data[(adult_data['gender'] == 'Female') & (adult_data['income'] == 1)].shape[0]

# Calculate the chances (proportion) for each gender
male_chances = males_above_50k / total_males
female_chances = females_above_50k / total_females

print(f"Chance of males earning >50k: {male_chances}")
print(f"Chance of females earning >50k: {female_chances}")
```
